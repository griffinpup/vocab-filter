```python
class S:
    
    def __init__(self, message, character):
        self.attributes = [-1] * len(attribute_index)
        self.message = message
        self.character = character
    
    def ChangeAttribute(self, value, idx):
        self.attributes[attribute_index[idx]] = value
    
    def ca(self, idx, value):
        self.ChangeAttribute(value, idx)
        
    def Print(self):
        print(self.message + " : " + self.character)
        #todo: print character
        
    
    
```

```python
class Vocab:
    def __init__(self):
        self.statements = []
        self.sorted = False
        
    def FilterSentence(input_sentence):
        return input_sentence
    
    def SortStatements(self):
        self.sorted_statements = []
        for value, key in enumerate(attribute_index):
            tmp_list = copy.deepcopy(self.statements)
            tmp_list.sort(reverse=True,key = lambda s : s.attributes[value])
            self.sorted_statements.append(tmp_list)
        self.IndexWords()
        self.sorted = True
    
    def IndexWords(self):
        self.indexed_words = {}
        for statement in self.statements:
            cur_words = statement.message.replace("?", "").replace(".", "").replace(",", "").replace("'", "").lower().split()
            for word in cur_words:
                if word in self.indexed_words:
                    if statement.character not in self.indexed_words[word]:
                        self.indexed_words[word].append(statement.character)
                else:
                    self.indexed_words[word] = [statement.character]
    
    def GetSentence(self, ideal_sentence):
        words = ideal_sentence.replace("?", "").replace(".", "").replace(",", "").replace("'", "").lower().split()
        potential_sentence = []
        potential_character = []
        ret_sentence = ""
        for word in words:
            if word in self.indexed_words:
                voices = ""
                for char in self.indexed_words[word].sort():
                    voices += "/" + char
                ret_sentence += word + " {" + voices + "} "
                potential_sentence.append(self.indexed_words[word])
            else:
                synonyms = GetSynonyms(word)
                for synonym in synonyms:
                    if synonym in self.indexed_words:
                        ret_sentence += synonym + " {" + self.indexed_words[synonym][0] + "} "
                        potential_character.append(self.indexed_words[synonym])
                        potential_sentence.append(synonym)
        
        character_count = {}
        for characters in potential_character:
            for character in characters:
                if character in character_count:
                    character_count[character] = character_count[character] + 1
                else:
                    character_count[character] = 1
            
            
        print(ret_sentence)
    
    def GetWord(self, word):
        if word in self.indexed_words:
            ret_sentence += word + " {" + self.indexed_words[word][0] + "} "
        
    def Take(self, attribute, count):
        if (self.sorted):
            for i in range(count):
                if i < len(self.sorted_statements[attribute_index[attribute]]):
                    print(self.sorted_statements[attribute_index[attribute]][i].Print())     
```

```python
def ParseParagraph(vocab_list, paragraph, character):
    sentences = paragraph.split("\n")
    for sentence in sentences:
        s = S(sentence, character)
        if "?" in sentence:
            s.ca("question", 10)
        vocab_list.statements.append(s)
```
